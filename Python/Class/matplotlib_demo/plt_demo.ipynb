```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8, 6])
ypoints = np.array([3, 10, 2])

plt.plot(xpoints, ypoints, "*")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
font = {'family': 'calibri','size': 20}
plt.xlabel("--->X",fontdict = font)
plt.ylabel("--->Y", fontdict = font)
plt.plot(xpoints, ypoints,"*")
plt.show()
```

# Default X-Points

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([2, 4, 8, 16, 32, 64])

plt.plot(ypoints,"*")
plt.show()
```
